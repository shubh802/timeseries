```python
## Optimization Function

def optimize_sarimax(endog, exog, parameters_list, d, D, s):
    """
        Returns df with parameters, corresponding AIC

        endog - observed variable
        exog - exogenous variables
        parameters_list - list of parameters to be optimized (p, d, P, Q) tuples
        d - integration order
        D - seasonal integration order
        s - length of the season
    """

    results = []

    for param in tqdm_notebook(parameters_list):
        try:
            model = SARIMAX(endog, exog=exog, 
                            order=(param[0], d, param[1]), 
                            seasonal_order=(param[2], D, param[3], s),
                            simple_differencing=False).fit(disp=False)
        except:
            continue

        aic = model.aic
        results.append((param, aic))

    result_df = pd.DataFrame(results, columns=['(p,q) x (P,Q)', 'AIC'])
    result_df = result_df.sort_values(by="AIC", ascending=True).reset_index(drop=True)

    return result_df
```

```python
p = range(0, 4, 1)
d = 1
q = range(0, 4, 1)
P = range(0, 4, 1)
D =  0 # There is no seasonal difference
Q = range(0, 4, 1)
s = 4 # length is 4 quarters

parameters = product(p, q, P, Q)
parameter_list = list(parameters)
print(parameter_list[:5])
```

```python
macro_data.columns
```

```python
# training with first 200
endog = macro_data.realgdp[:200]
exog = macro_data[['realcons', 'realinv', 'realgovt', 'realdpi', 'cpi', 'm1', 'tbilrate', 'unemp', 'pop', 'infl', 'realint']][:200]

result_df = optimize_sarimax(endog, exog, parameter_list, d, D, s)
result_df.head(5)
```

```python
best_model = SARIMAX(endog, exog=exog, order=(2, 1, 2), 
                    seasonal_order=(1, 0, 0, 4), 
                    simple_differencing=False)
res = best_model.fit(disp=False)

print(res.summary())
```

```python
res.plot_diagnostics();
```

```python
n_forecast = 3
predict = res.get_prediction(end = best_model.nobs + n_forecast,
                            exog = exog.iloc[-4:]) # to give us the last 3 exog observation as to predict the next 3

idx = np.arange(len(predict.predicted_mean))
predict_ci = predict.conf_int(alpha=0.5)

fig, ax = plt.subplots()
ax.plot(macro_data.realgdp[:199], color='blue', linewidth=1, label='Real GDP')
ax.plot(idx[-n_forecast:], predict.predicted_mean[-n_forecast:], 'k--', linewidth=1, label='Predicted GDP')
ax.fill_between(idx, predict_ci.iloc[:,0], predict_ci.iloc[:,1], color='red', alpha=0.2)

ax.set_title("Forecast Of US Real GDP")
plt.legend()
plt.show()
```

```python
## Zooming in to see the Conditional Interval
fig, ax = plt.subplots()
ax.plot(macro_data.realgdp, color='blue', linewidth=1, label='Real GDP')
ax.plot(idx[-n_forecast:], predict.predicted_mean[-n_forecast:], 'k--', linewidth=1, label='Predicted GDP')
ax.plot(predict.predicted_mean, 'ro')
ax.fill_between(idx, predict_ci.iloc[:,0], predict_ci.iloc[:,1], color='red', alpha=0.2)
ax.set_xlim(190, 199)
ax.set_ylim(13000, 14100)

ax.set_title("Forecast Of US Real GDP")
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(macro_data.realgdp, predict.predicted_mean[:203])

print(f"MSE: {mse}")
```

By looking at the graph we could see the model is working fine but from mse metrics we are not sure, so we would look into the mean percentage error as the metric

```python
mape = np.mean(np.abs((macro_data.realgdp - predict.predicted_mean[:203]) / macro_data.realgdp))

print(f"MAPE: {round(mape * 100, 2)}%")
```
